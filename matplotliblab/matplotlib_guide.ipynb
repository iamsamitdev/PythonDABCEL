#### 1. Import Matplotlib

```python
import matplotlib.pyplot as plt
```

#### 2. Plot a line graph

```python
import numpy as np

# กำหนดค่าของแกน x และ y
x = np.array([2018, 2019, 2020, 2021, 2022])
y = np.array([3000, 2500, 4200, 600, 1300])

# plot กราฟด้วย matplotlib
plt.plot(x, y)

# แสดงผลกราฟ
plt.show()
```

#### 3. Plot from CSV file

```python
import pandas as pd

df = pd.read_csv('dataset/gdp.csv')

plt.plot(df['years'], df['gdp'], marker='.', markersize=10, mfc='g', linestyle='dotted') # dotted, dashed, dashdot, solid, None
plt.title('Thailand GDP', c='b')
plt.xlabel('Year', c='b')
plt.ylabel('GDP', c='b')

plt.show()
```

#### 3. Plot Bar Chart

```python
import pandas as pd

df = pd.read_csv('dataset/gdp.csv')

plt.bar(df['years'], df['gdp'], color=['r', 'g', 'b', 'y', 'm', 'c', 'k', 'w'])
plt.title('Thailand GDP' , c='b')
plt.xlabel('Year', c='b')
plt.ylabel('GDP', c='b')

plt.show()
```

#### 4. Plot Scatter Chart

```python
import pandas as pd

df = pd.read_csv('dataset/gdp.csv')

plt.scatter(df['years'], df['gdp'], c='r', marker='o', s=50)
plt.title('Thailand GDP' , c='b')
plt.xlabel('Year', c='b')
plt.ylabel('GDP', c='b')

plt.show()
```

#### 5. Plot Pie Chart

```python
import pandas as pd

df = pd.read_csv('dataset/gdp.csv')

plt.pie(df['gdp'], labels=df['years'], autopct='%.2f%%')
plt.title('Thailand GDP' , c='b')
plt.xlabel('Year', c='b')
plt.ylabel('GDP', c='b')

plt.show()
```
